```python
def query(sql):
    res = pandas.read_sql_query(sql,db,chunksize=100_000)
    return next(res)
```

```python
res = query("SELECT gender, usertype, count(*) from tripdata group by gender, usertype")
res
```

```python
res = query("""
 SELECT (2020-"birth year") as age,
 count(*)
 from tripdata
 group by age
 having age > 1
""")
res
```

```python
res = query("""
 SELECT bikeid, sum(tripduration)
 from tripdata
 group by bikeid
 ORDER by 2
""")
res
```

```python
res = query("""
 SELECT gender, sum(tripduration)
 from tripdata
 group by gender
""")
res
```

```python
res = query("""
 SELECT bikeid, sum(tripduration) as total
 from tripdata
 group by bikeid
""")
res
```

```python
query("""
select hour_bucket, count(*) from
(
 SELECT bikeid, sum(tripduration)/3600 as hours, round(sum(tripduration) / 3600 / 100) * 100  as hour_bucket
 from tripdata
 group by bikeid
 ) 
 group by hour_bucket
""")
```

```python
query("""
select starttime, substr(starttime, 1, 10) as date, substr(starttime, 1, 4) as y, substr(starttime, 6, 2) as m,
substr(starttime, 9, 2) as d, substr(starttime, 11, 10) as t
from tripdata
limit 3
""")
```

```python
query("""
select bikeid, substr(starttime, 6, 2) as m,
sum(tripduration)/3600 as total
from 
tripdata
group by
bikeid, m
order by total
""")
```

```python
query("""
select
  bikeid,
  sum(case when m = '10' then total end) as oct
  from
  (
select bikeid, substr(starttime, 6, 2) as m,
sum(tripduration)/3600 as total
from 
tripdata
group by
bikeid, m
order by total
)
group by bikeid
""")
```

```python
query("""
select
  bikeid,
  sum(case when m = '10' then total end) as oct,
  sum(case when m = '11' then total end) as nov,
  sum(case when m = '12' then total end) as dec
  from
  (
select bikeid, substr(starttime, 6, 2) as m,
sum(tripduration)/3600 as total
from 
tripdata
group by
bikeid, m
order by total
)
group by bikeid
""")
```
